```python
import requests
import pandas as pd
import time
from datetime import datetime
from sklearn.linear_model import LinearRegression
```

```python
def Quote_change(today,yesterday):
    change = (today-yesterday)/yesterday
    return change
```

```python
model = LinearRegression()
tokens = ""
```

```python
all_0050 = pd.read_csv("0050_all_stock.csv")
```

```python
b = []

for i in range(0,all_0050.shape[0]):#all_0050.shape[0]
    a=[]
    url = "https://api.finmindtrade.com/api/v4/data"
    parameter = {
        "dataset": "TaiwanStockPrice",
        "data_id": "{}".format(all_0050.iloc[i,1]),
        "start_date": "2021-03-01",
        "end_date": "2021-03-31",
        "token": tokens, # 參考登入，獲取金鑰
    }
    #time.sleep(3)
    a.append(parameter["data_id"])
    
    resp = requests.get(url, params=parameter)    
    data = resp.json()
    data = pd.DataFrame(data["data"])
    #print(data.shape)
    print("data_id: ","{}".format(all_0050.iloc[i,1]))
    date = (data["date"][-10:])
    
    close_yesterday = list(data["close"])
    close_yesterday.insert(0,"num")         #在第一行塞0讓他們的位置對齊
    close_yesterday = close_yesterday[:-1]  #因在第一行補0，那最後一個收盤價相對會被往下移，透過此行刪除最後一個收盤價
    close_yesterday = close_yesterday[-10:]
    close_yesterday    
    
    close_ = data["close"]
    del data["stock_id"]
    del data["date"]
    
    X_train = data.iloc[:-11,:]  #以倒數最後10天以前的資料作訓練
    y_train = close_[1:-10]
    X_test = data.iloc[-11:-1]
    y_test = close_[-10:]


    model.fit(X_train, y_train)
    y_test_hat = model.predict(X_test)
    y_test_hat = y_test_hat.round(4)  #取小數點後四位
    
    predition_data = []
    for k in range(0,10):        #抓後10天的資料
        g = []
        g.append(date.iloc[k])
        g.append(y_test_hat[k])
        g.append(y_test.iloc[k])
        g.append(Quote_change(y_test_hat[k],close_yesterday[k]))
        
        
        predition_data.append(g)
    a.append(predition_data)
    

    b.append(a)
    
    now = datetime.now()                          #計算每次迴圈的運算時間，並且可以確保迴圈還在運行
    current_time = now.strftime("%H:%M:%S")
    print("Current Time =", current_time)
```

```python
predition_date  = []
for z in range(0,len(b[0][1])):
    predition_date.append(b[0][1][z][0])
```

```python
by_date=[]
for i in range(0,len(predition_date)):
    for n in range(0,len(b)):
        for m in range(0,len(b[0][1])):
            if predition_date[i] == b[n][1][m][0]:   #紀錄時間
                kk = []
                kk.append(predition_date[i])
                kk.append(b[n][0])
                kk.append(b[n][1][m][3])
                print(predition_date[i],b[n][0],b[n][1][m][3])
                by_date.append(kk)
```

```python
by_date_data = pd.DataFrame(by_date)
pd.DataFrame(by_date_data).columns=["date","stock_id","updown"]  #命名
```

```python
rank = list(range(1,50+1))
```

```python
b = pd.DataFrame()
for k in range(0,len(predition_date)):
    fliter = by_date_data["date"] == predition_date[k]
    a = by_date_data[fliter].sort_values(by=['updown'], ascending=False)
    a["rank"] = rank
    b = pd.concat([b,a], ignore_index=True)
del b["updown"]
```

```python
b.sort_values(by=["date","stock_id"], inplace=True)
b= b.reset_index(drop=True)    #把index重設
b = pd.concat([pd.Series(b.index),b["stock_id"],b["date"],b["rank"]],axis=1)
b.columns.values[0] = "index"
b.head()
```

```python
LinearRegression_index_etc=[]
for j in range(0,len(b)):
    LinearRegression_index_etc.append(str(b["index"][j])+" ; "+str(b["date"][j])+" ; "+str(b["stock_id"][j]))

LinearRegression = pd.DataFrame()
LinearRegression["index ; date ; stock_id"] = LinearRegression_index_etc
LinearRegression["rank"] = b["rank"]/50
```

```python
LinearRegression.to_csv("LinearRegression.csv",index=False)
```
